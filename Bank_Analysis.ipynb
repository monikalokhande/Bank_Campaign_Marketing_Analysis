### que- Which customer profiles (job, education, etc.) have higher subscription rates?

```python
b=df.groupby('job')['y'].value_counts(normalize=True).unstack()*100
```

```python
b.plot(kind='bar',figsize=(10,5),color=['Skyblue','Brown'])
plt.xticks(rotation=90)
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title("Subscription Rates by Customer Job Profile")
plt.xlabel("Job Profile")
plt.ylabel("Subscription Rate (%)")
plt.show()
```

```python
a=df.groupby('education')['y'].value_counts(normalize=True).unstack()*100
```

# Campaign Effectiveness:

### que - Does the duration of the call increase the chances of a successful subscription?

```python
c=df.groupby('y')['duration'].mean().reset_index()
```

```python
c
```

```python
sb.boxplot(x=df.y,y=df.duration)
plt.title("Impact of Call Duration on Subscription Success Rate")
plt.xlabel("Subscription Success Rate (%)")
plt.ylabel("Call Duration (sec)")
plt.show()
```

### que - Which contact methods (e.g., telephone, cellular) are most effective?


```python
d=df.groupby('contact')['y'].value_counts(normalize=True).unstack()
```

```python
d
```

```python
d.plot(kind='bar',figsize=(10,5),color=['grey','yellow'])
plt.xticks(rotation=90)
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.15,1.02))
plt.title("Effectiveness of Different Contact Methods")
plt.xlabel("Contact Methods")
plt.ylabel("Effectiveness Rating")
plt.show()
```

### que - How many contacts are usually needed before a client subscribes?

```python
e=df.groupby('y')['campaign'].mean().reset_index()
```

```python

sb.boxplot(x=df.y,y=df.campaign)
plt.title("Average Number of Contacts Before Client Subscription")
plt.xlabel("Number of Contacts Made")
plt.ylabel("Number of Clients Who Subscribed")
plt.show()
```

# Previous Campaign Impact:

### que - How does the outcome of previous campaigns influence the current one?

```python
f=df.groupby('poutcome')['y'].value_counts(normalize=True).unstack()
```

```python

f.plot(kind='bar',figsize=(10,5),color=['skyblue','yellow'])
plt.xticks(rotation=90)
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.15,1.02))
plt.title("Impact of Previous Campaign Outcomes on Current Campaign Success")
plt.xlabel("Outcome of Previous Campaign")
plt.ylabel("Current Campaign Performance Metric")
plt.show()
```
